```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt


```

```python
data = pd.read_csv('loan_app_dataset.csv')
data.head()
```

```python
df = data.copy()
loan_status_counts = df[' loan_status'].value_counts()
plt.figure(figsize=(6,4))
plt.bar (loan_status_counts.index, loan_status_counts.values, color = ['orange', 'green'])
plt.title('Loan Status Distribution')
plt.ylabel('count')
plt.show()
```

```python
x = (loan_status_counts.index)
y = (loan_status_counts.values)
plt.bar(x,y)

plt.title ('Graph')
plt.xlabel(loan_status_counts.index)
plt.ylabel('count')
plt.show()
```
